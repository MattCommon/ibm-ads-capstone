```python
import pickle
import numpy as np
from sklearn.utils import shuffle
```

```python
# Define function to read in pickled data from etl step
def readData(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)
```

```python
# Read in training data
X_train = readData('X_train.pickle')
y_train = readData('y_train.pickle')
images_train = readData('images_train.pickle')
labels_train = readData('labels_train.pickle')
```

```python
# Read in test data
X_test = readData('X_test.pickle')
y_test = readData('y_test.pickle')
images_test = readData('images_test.pickle')
labels_test = readData('labels_test.pickle')
```

```python
# Shuffle training datasets
X_train, y_train, images_train, labels_train = shuffle(X_train, y_train, images_train, labels_train)
```

# Simple Feed Forward Network

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
batch_size = 128
num_classes = 6
epochs = 20
inputSize = 150 * 150 * 3
inputSize
```

```python
X_train = X_train.reshape(14034, inputSize)
X_test = X_test.reshape(3000, inputSize)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(512, activation='relu',
input_shape=(inputSize,))) # First layer only
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=batch_size,
          epochs=epochs, validation_data=(X_test, y_test))

score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

# Convolutional Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPool2D
```

```python

X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
X_train.shape
```

```python
X_train
```

```python
epochs = 10
#create model
model = Sequential()
#add model layers
model.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(150, 150, 3)))
model.add(Conv2D(64, kernel_size=3, activation='relu'))
model.add(MaxPool2D(3, 3))
model.add(Conv2D(128, kernel_size=3, activation='relu'))
model.add(Conv2D(128, kernel_size=3, activation='relu'))
model.add(MaxPool2D(3, 3))
model.add(Conv2D(256, kernel_size=3, activation='relu'))
model.add(Conv2D(256, kernel_size=3, activation='relu'))
model.add(Conv2D(256, kernel_size=3, activation='relu'))
model.add(MaxPool2D(3, 3))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(6, activation='softmax'))
model.summary()
```

```python

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train, batch_size=batch_size,
          epochs=epochs, validation_split=0.3)

score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
